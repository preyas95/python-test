```python

from nsepy import get_history
from datetime import date
import pandas as pd
#from pandas.tseries.offsets import *
stock=input("Choose stocks:\t TCS or INFY \n")
data = get_history(symbol=stock, start=date(2015,1,1), end=date(2015,12,31))
#data[['Close']].plot()
#data.to_csv('/home/preyas/Downloads/xxx.csv')
#data = pd.read_csv('/home/preyas/Downloads/xxx.csv')
#print(data.head())


weeks=input("Weeks 4,8,16,...52\n")
weeks = int(weeks)
ndays=weeks*7 
 
def MA(ndays,data):
    SMA=pd.Series((data['Close']).rolling(window=ndays).mean(),name = 'SMA')
    data=data.join(SMA)
    data.to_csv('/home/preyas/Downloads/xxx.csv')
    return(data)

data=MA(ndays,data)
```
